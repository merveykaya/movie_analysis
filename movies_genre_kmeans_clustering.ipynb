## Data Attributes

|Attribute|Description|
|:---------:|:-----------:|
|popularity|	Each model builds their popularity value slightly differently. |budget| The money spend on production process|
|revenue| Earned money from the film|

|director|	The director of the film|
|genres| All genres accepted for the film|
